**Aim** : To implement the chunking with NLTK Toolkit <br/>

```python
import nltk
```

```python
nltk.download('punkt')
```

```python
nltk.download('averaged_perceptron_tagger')
```

```python
sentence = "the little white dog barked at the crowd"
```

```python
grammar = ("NP: {<DT>?<JJ>*<NN>}")
```

```python
chunkParser = nltk.RegexpParser(grammar)
tagged = nltk.pos_tag(nltk.word_tokenize(sentence))
print(tagged)
```

```python
tree = chunkParser.parse(tagged)
```

```python
for subtree in tree.subtrees():
    print(subtree)
```
